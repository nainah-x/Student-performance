## Dataset Overview

> The Students Performance dataset originally contained 1000 rows and 8 variables (columns) consisting of both categorical and numerical variables. I created additional 2 columns which included the average score and total score of the students in the test. The dataset was sourced from kaggle.

> The dataset contains the following columns:

1. Gender: Male/Female.
2. race/ethnicity: Ethnic group A,B,C or D.
3. parental level of education: High school, some High school, College, Bachelors' degree, Associate degree,Master's degree.
4. lunch: standard/reduced.
5. test preparation course: described whether studnts completed the test preparation course or not.
6. math score: describes students score in maths out of 100%.
7. reading score: describes students score in reading out of 100%.
8. writing score: describes students score in writing out of 100%. 
9. average_score: an average of the total scores of students in all 3 subjects.
10. total_score: an addition of all the students scores combined together.


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df=pd.read_csv('StudentsPerformance.csv')
```

Majority of the students that scored between 80-100%; the 'Top scorers' had Bachelor's degree holding parents, followed by master's degree and associate's degree. Parental level of education did have a significant influence on the students' performance in tests. 

```python
df['average_score'] = df.mean(numeric_only=True, axis=1)
df['average_score'] =df['average_score'].apply(lambda x: round(x, 0))
def x_y_z(xaxix,yaxis,title):
    plt.xlabel(xaxix)
    plt.ylabel(yaxis)
    plt.title(title)
df['parental level of education']=pd.Categorical(df['parental level of education'], categories=['some high school', 'high school', 'some college', "associate's degree", "bachelor's degree", "master's degree"],
                                   ordered=True)
# plot the graph
plt.figure(figsize=[14, 6])
top_scorers = df[df['average_score'] > 80].sort_index()
ax=sb.violinplot(x='average_score', y='parental level of education', data=top_scorers, palette='Dark2_r');
x_y_z('Average Score', 'Parental Level Of Education', 'Effect Of Parental Level Of Education On Students Performance')
```

Studies have found that access to nutrition, particularly breakfast not only enhances students performance but can also enhance students psychosocial well-being, reduce aggression and school suspensions, and decrease discipline problems. This is infact true as shown in this data, students that had standard lunch performanced better overall in the tests than the students who had free/reduced lunch. 

```python
plt.figure(figsize=[14, 6])
sb.boxplot(x='average_score', y='lunch', data=df);
x_y_z('Average score', 'Lunch', 'Effect of Lunch Type on Students Average Score');
```

Parents of students from ethnic group E have the highest level of Educational background, having the highest rate of associate's degree, bachelor's degree and master's degree. And of course their wards had the highest performance in the test. This goes on to show the influnce of ethnicity on parental level of education and how that subsequently affects the students academic performance.



```python
df['race/ethnicity']=pd.Categorical(df['race/ethnicity'], categories=['group A', 'group B', 'group C', 'group D', 'group E'],
                                   ordered=True)
plt.figure(figsize=[14, 6])
sb.boxplot(x='average_score', y='race/ethnicity', data=df, palette='Blues');
x_y_z('average score', 'race/ethnicity', 'Effect of Race/ethnicity on Students Average Score');
```
